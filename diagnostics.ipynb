# wflow diagnostics

```python
import numpy as np
import pylab as p

from parse_wflow import strip_blocks
from diagnostics import inname, convert, difference

%matplotlib inline
# Update the matplotlib configuration parameters:
p.rcParams.update({'font.size': 11, 'font.family': 'serif'})
```

```python
def plot_deltas(deltas, ylim = None):
    "Plot difference of RK integration schemes."
    p.xlabel('t')
    p.ylabel('$\Delta [t^2 E]$')
    color = ['b', 'r', 'm', 'k', 'g']
    dt = [.1,.05,.025,.01,.005]
    for i in range(len(deltas)-1):
        t = blocks[i][:,0]
        if i % 2 == 0:
            f = '--'+color[i/2]
            l = 'dt={0}'.format(dt[i/2])
        else:
            f = '-'+color[i/2]
            l = None
        p.errorbar(t, deltas[i], fmt=f, label=l)
    p.legend(loc=0)
    if ylim:
        p.ylim(ylim)
    p.show()
```

```python
with open(inname, 'r') as f:
    dat = f.readlines()
blocks = strip_blocks(dat, 8)
blocks = map(np.array, blocks)
ds = [convert(b[:,0], b[:,2]) for b in blocks]
deltas = [difference(d, ds[-1]) for d in ds]
```

### Dependence of observables on $\Delta t$ and integration scheme.

Below shows the difference in $t^2 E$ flow (on a single config) as the time step is varied.  Dotted lines are Euler integrator, while solid lines are RK4 integrator. Colors denote timestep as in the legend.  The "reference" flow is RK4 with dt=0.005.  The second plot zooms into the axis, where some deviations with RK4 at larger timesteps become apparent.

```python
p.figure(figsize=(8,4))
plot_deltas(deltas)
p.figure(figsize=(8,5))
plot_deltas(deltas,[-.001,.001])
```
