# Wilson Flow results

```python
import numpy as np
import pylab as p

from parse_wflow import file2numpy
from analyze_wflow import inname, plot_items, dfb
from resample import JK_block, JKsigma
```

```python
def plot_flow(t2E):
    nstep = len(t2E[0])
    t = np.linspace(0,1.0,nstep)
    p.xlabel('step')
    p.errorbar(range(nstep),[0.3]*nstep, fmt='k--')
    p.errorbar(range(nstep), t2E[0], label='t^2<E>')
    p.errorbar(range(nstep), 100*t*dfb(t2E[0]), label='t d/dt [t^2<E>]')
    p.legend(loc='best')
    p.show()
```

## 16^4, Beta = 5.2875, am=0.025

```python
dat1 = file2numpy(inname(16,5.2875,0.025))
```

```python
nconf, nstep, nmeas = dat1.shape
p.ylabel('t^2 E')
plot_items([dat1[x,:,2] for x in range(1,nconf)])
```

```python
t2E = JK_block(dat1[1:,:,2])  # I threw out the first config.
plot_flow(t2E)
```

## 16^4, Beta=5.4, am=0.025

```python
dat = file2numpy(inname(16, 5.4, 0.025))
nconf, nstep, nmeas = dat.shape
p.ylabel('t^2 E')
plot_items([dat[x,:,2] for x in range(1,nconf)])
```

```python
t2E = JK_block(dat[:,:,2])
plot_flow(t2E)
```
