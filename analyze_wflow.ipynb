# Wilson Flow results

NOTE: In this notebook $t_0/a$ is dimensionless.  In other words my $t_0$ is equivalent to $\sqrt{t_0}$ in the literature, and has the same dimensions as $w_0$.
+ 15-Aug: Overall normalization added: $E\rightarrow \frac{3}{2}E$
+ 15-Aug: Error bars added to plots.

-----

```python
import numpy as np
import pylab as p
from scipy.interpolate import interp1d

from parse_wflow import file2numpy
from analyze_wflow import inname, plot_items, dfb, broot
from resample import JK_block, JKsigma
```

```python
%matplotlib inline
# Update the matplotlib configuration parameters:
p.rcParams.update({'font.size': 11, 'font.family': 'serif'})
```

```python
def plot_flow(t2E):
    dt = 0.01
    nstep = len(t2E[0])
    t = np.linspace(0,nstep*dt,nstep)
    p.xlabel('step')
    p.xlim([0,nstep])
    y, err = t2E[0], JKsigma(t2E)
    p.errorbar(range(nstep), y, label='t^2<E>')
    p.fill_between(range(nstep), y+err, y-err, color='b', alpha=0.5)
    y, err = t*dfb(t2E[0])/dt, JKsigma(t*dfb(t2E)/dt)
    p.errorbar(range(nstep), y, label='t d/dt [t^2<E>]')
    p.fill_between(range(nstep), y+err, y-err, color='g', alpha=0.5)
    p.legend(loc='best')
    p.errorbar(range(nstep),[0.3]*nstep, fmt='k--')
    p.show()
```

```python
C = 1.5  # Normalization factor for E.
dt = 0.01
```

### 16^4, Beta = 5.2875, am=0.025

```python
dat1 = file2numpy(inname(16,5.2875,0.025))
nconf, nstep, nmeas = dat1.shape
p.ylabel('t^2 E')
plot_items([dat1[x,:,2] for x in range(1,nconf)])
```

```python
t2E = C*JK_block(dat1[1:,:,2])  # I threw out the first config.
nconfig, nstep = t2E.shape
t = np.linspace(0, nstep*dt, nstep)
f=interp1d(t,t2E[0]-0.3, kind='cubic')
g=interp1d(t, t*dfb(t2E[0])/dt - 0.3, kind='cubic')
t02 = broot(f, 0.0, nstep*dt)
w02 = broot(g, 0.0, nstep*dt)
p.vlines([t02/dt, w02/dt], 0, 0.3, colors='k', linestyles='dotted')
plot_flow(t2E)
t0 = np.sqrt(t02)
w0 = np.sqrt(w02)
print 't0/a = {0}'.format(t0)
print 'w0/a = {0}'.format(w0)
```
