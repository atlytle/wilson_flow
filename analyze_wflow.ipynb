# Wilson Flow results

NOTE: In this notebook $t_0/a$ is dimensionless.  In other words my $t_0$ is equivalent to $\sqrt{t_0}$ in the literature, and has the same dimensions as $w_0$.
+ 15-Aug: Overall normalization added: $E\rightarrow \frac{3}{2}E$
+ 15-Aug: Error bars added to plots.
+ 03-Sep: Run 2 results.  Includes clover determinations.

-----

```python
import numpy as np
import pylab as p
from scipy.interpolate import interp1d

from parse_wflow import file2numpy
from analyze_wflow import inname, plot_items, broot
from analyze_wflow import dfb2 as d
from resample import JK_block, JKsigma
```

```python
%matplotlib inline
# Update the matplotlib configuration parameters:
p.rcParams.update({'font.size': 11, 'font.family': 'serif'})
```

```python
def plot_flow(t2E):
    dt = 0.01
    nstep = len(t2E[0])
    t = np.linspace(0,nstep*dt,nstep)
    p.xlabel('step')
    p.xlim([0,nstep])
    y, err = t2E[0], JKsigma(t2E)
    p.errorbar(range(nstep), y, label='t^2<E>')
    p.fill_between(range(nstep), y+err, y-err, color='b', alpha=0.5)
    y, err = t*d(t2E[0])/dt, JKsigma(t*d(t2E)/dt)
    p.errorbar(range(nstep), y, label='t d/dt [t^2<E>]')
    p.fill_between(range(nstep), y+err, y-err, color='g', alpha=0.5)
    p.legend(loc='best')
    p.errorbar(range(nstep),[0.3]*nstep, fmt='k--')
    p.show()
    
def plot_flows(t, t2Ep, t2Ec):
    dt = 0.01
    p.xlabel('t')
    
    y = t2Ep[0]
    p.errorbar(t, y, label='t^2<E_p>')
    
    y = t*d(t2Ep[0])/dt
    p.errorbar(t, y, fmt='b--')
    
    y = t2Ec[0]
    p.errorbar(t, y, label='t^2<E_c>')
    
    y = t*d(t2Ec[0])/dt
    p.errorbar(t, y, fmt='g--')
    
    p.errorbar(t, [0.3]*len(t), fmt='k:')
    
    p.legend(loc='best')
    
    p.show()

def ferr((x, dx), n):
    "Format number to precision n in x(dx) form."
    fs = '{0:.'+str(n)+'f}({1})'  # Format string.
    dx *= (10**n)
    dx = int(round(dx))
    if dx == 0:
        dx = 1
    return fs.format(x, dx)

ferr3 = lambda x: ferr(x,3)
ferr4 = lambda x: ferr(x,4)

def find_root(t, y):
    f = interp1d(t, y, kind='cubic')
    return broot(f, t[0], t[-1])  # Assumes unique root.

def JKroot(t, ys):
    fr = lambda y: np.sqrt(find_root(t,y-0.3))  # Note sqrt() here.
    return np.array(map(fr, ys))
    #return vals[0], JKsigma(vals)
    
jkvals = lambda vals: (vals[0], JKsigma(vals))

def find_roots(t, t2Ep, t2Ec, div=False):
    "Various flow times."
    t0p = JKroot(t, t2Ep)
    t0c = JKroot(t, t2Ec)
    w0p = JKroot(t, t*d(t2Ep)/dt)
    w0c = JKroot(t, t*d(t2Ec)/dt )
    if div: 
        return t0p/t0c, w0p/w0c, w0c/t0c, w0p/t0p
    
    return t0p, t0c, w0p, w0c 
```

```python
C = 1.5  # Normalization factor for E.
dt = 0.01
```

16^4 runs
----------

### 16^4, $\beta$ = 5.2875, am=0.015 Run 2

```python
dat = file2numpy(inname(16, 5.2875, 0.015, 'b'), offset=9)
nconf, nstep, nmeas = dat.shape
p.ylabel('t^2 E')
plot_items([dat[x,:,1] for x in range(nconf)])
```

```python
t = dat[0,:,0]
t2Ep = JK_block(dat[:,:,1])  # Plaquette.
t2Ec = JK_block(dat[:,:,3])  # Clover.
t0p, t0c, w0p, w0c = map(jkvals, find_roots(t, t2Ep, t2Ec))

print ' t0p = {0}\n t0c = {1}\n w0p = {2}\n w0c = {3}'.format(*map(ferr4, [t0p, t0c, w0p, w0c]))
plot_flows(t, t2Ep, t2Ec)
```
